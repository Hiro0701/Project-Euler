```python
def fibonacci_sequence(n: int) -> list:
    fibonacci_list = [1, 2]
    
    if n == 1:
        return [1]
    elif n == 2:
        return [1, 2]
    else:
        for i in range(3, n+1):
            fibonacci_list.append(fibonacci_list[-1] + fibonacci_list[-2])
        return fibonacci_list        
```

```python
n = 1

while True:
    if fibonacci_sequence(n)[-1] > 4000000:
        fibonacci_n = fibonacci_sequence(n)[:-1]
        break

    n += 1

print(fibonacci_n)
```

```python
fibonacci_even_sum = sum(
    [n for n in fibonacci_n if n % 2 == 0]
)

print(fibonacci_even_sum)
```

```python
def fibonacci_even_sum_sol(limit: int) -> int:
    a = 0
    b = 1
    c = 2

    sum = 2

    while c < limit:
        sum += c

        a = b + c
        b = a + c
        c = a + b

    return sum
```

```python
sol = fibonacci_even_sum_sol(4000000)

print(sol)
```
