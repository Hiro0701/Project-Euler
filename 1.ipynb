```python
def build_set(n) -> set:
    return set(range(1,n+1))

def check_multiple(natural_numbers: set, n: int) -> set:
    return set(
        i for i in natural_numbers if i % n == 0
    )
```

```python
set1 = build_set(9)

set1
```

```python
set1_3 = check_multiple(set1, 3)
set1_5 = check_multiple(set1, 5)
set1_3_5 = set1_3 | set1_5
```

```python
print(set1_3_5)

print(sum(set1_3_5))
```

```python
set2 = build_set(999)

set2_3 = check_multiple(set2, 3)
set2_5 = check_multiple(set2, 5)
set2_3_5 = set2_3 | set2_5
```

```python
print(set2_3_5)

print(sum(set2_3_5))
```
