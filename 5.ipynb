```python
def smallest_common_multiple(n: int) -> int:
    a = 6
    b = 0
    q = 0
    r = 0
    temp = 0
    gcd = 1
    lcm = 1

    for i in range(4, n+1):
        b = a
        q = i
        while True:
            r = b % q
            if r == 0:
                gcd = q
                lcm = (a * i) // gcd
                break
            b = q
            q = r
        
        a = lcm
    
    return lcm
```

```python
common_multiple_10 = smallest_common_multiple(10)

print(common_multiple_10)
```
